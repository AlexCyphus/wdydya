```python
def analyse(dataframe):

    #Read and split file into Android + iOS
    filename = dataframe + ".xlsx"
    df = pd.read_excel(filename)

    is_android = df.iloc[:,36] == "Android"
    df_android = df[is_android]

    is_ios = df.iloc[:,36] != "Android"
    df_ios = df[is_ios]

    # Count number of device specific responses
    glb.all_android += df_android["#"].count()
    glb.all_ios += df_ios["#"].count()
    glb.lang_total += df["#"].count()
    
    # Counting specific entries for entire dataframe (for this language)

    # 1 My conversations on Tandem rarely developed into a meaningful conversation
    b = df.iloc[:, 1]
    glb.convo_undeveloped += b.count()

    # 2 The app is not what I expected
    c = df.iloc[:, 2]
    glb.not_expected += c.count()

    # 3 There were too many scammers and fake profiles
    d = df.iloc[:,3]
    glb.scammers += d.count()

    # 4 I couldn't find anyone to talk to
    e = df.iloc[:,4]
    glb.no_partner += e.count()

    # 5 The app didn't help me much with my language learning
    f = df.iloc[:,5]
    glb.no_help += f.count()

    # 6 I received too many messages on Tandem
    g = df.iloc[:,6]
    glb.too_many_msgs += g.count()

    # 7 I received too many calls on Tandem that I didn't want to take
    h = df.iloc[:,7]
    glb.too_many_calls += h.count()

    # 8 Someone offended or annoyed me
    i = df.iloc[:,8]
    glb.offended += i.count()

    # 9 I had technical problems using the app

    # Total across iOS + Android
    j = df.iloc[:,9]
    glb.technical += j.count()

    # Seperated by iOS and Android
    j_android = df_android.iloc[:,9]
    glb.technical_android += j_android.count()
    j_ios = df_ios.iloc[:,9]
    glb.technical_ios += j_ios.count()

    # 10 The app takes too much space on my device

    # Total across iOS + Android
    k = df.iloc[:,10]
    glb.too_much_space += k.count()

    # Seperated by iOS and Android
    k_android = df_android.iloc[:,10]
    glb.too_much_space_android += k_android.count()
    k_ios = df_ios.iloc[:,10]
    glb.too_much_space_ios += k_ios.count()

    # 11 The app drains my battery

    # Total across iOS + Android
    l = df.iloc[:,11]
    glb.drains_battery += l.count()

    # Seperated by iOS and Android
    l_android = df_android.iloc[:,11]
    glb.drains_battery_android += l_android.count()
    l_ios = df_ios.iloc[:,11]
    glb.drains_battery_ios += l_ios.count()

    # 12 The app sent me too many push notifications, it got annoying

    # Total across iOS + Android
    m = df.iloc[:,12]
    glb.push_notifications += m.count()

    # Seperated by iOS and Android
    m_android = df_android.iloc[:,12]
    glb.push_notifications_android += m_android.count()
    m_ios = df_ios.iloc[:,12]
    glb.push_notifications_ios += m_ios.count()

    # 13 The app was confusing, I didn't understand how to use it
    n = df.iloc[:,13]
    glb.confusing += n.count()

    # 14 I'm not learning a language anymore
    o = df.iloc[:,14]
    glb.not_learning += o.count()

    # 15 I don't have enough time for the app
    p = df.iloc[:,15]
    glb.no_time += p.count()

    # 16 I want to change my profile information
    q = df.iloc[:,16]
    glb.change_profile += q.count()

    # 17 I want to start a new account
    r = df.iloc[:,17]
    glb.new_account += r.count()

    # 18 I'm a beginner and don't speak well enough yet to have a proper language exchange
    s = df.iloc[:,18]
    glb.beginner += s.count()

    # 19 It felt like a dating app
    t = df.iloc[:,19]
    glb.dating += t.count()

    # 20 I don't want to pay for Tandem and the free features are not enough for me
    u = df.iloc[:,20]
    glb.no_pay += u.count()

    # 21 I'm using a different app instead of Tandem
    v = df.iloc[:,21]
    glb.different_app += v.count()

    # 22 My conversations were almost exclusively in one language
    w = df.iloc[:,22]
    glb.one_lang += w.count()

    # 23 Other (reason)
    x = df.iloc[:,23]
    glb.other_reason += x.count()

    # * 24 You mentioned that you're using a different app instead of Tandem. Which app are you using?
    y = df.iloc[:,24]

    glb.hellotalk += y[(y == "HelloTalk") == True].count()
    glb.speaky += y[(y == "Speaky") == True].count()
    glb.fylp += y[(y == "Fylp") == True].count()
    glb.italki += y[(y == "iTalki") == True].count()
    glb.opentalk += y[(y == "Opentalk") == True].count()
    glb.lingbe += y[(y == "Lingbe") == True].count()
    glb.bilingua += y[(y == "Bilingua") == True].count()
    glb.hellopal += y[(y == "Hello Pal") == True].count()
    glb.duolingo += y[(y == "Duolingo") == True].count()
    glb.ablo += y[(y == "Ablo") == True].count()
    glb.interpals += y[(y == "InterPals") == True].count()

    # 25 Other (app)
    z = df.iloc[:,25]
    glb.other_app += z.count()

    # 26 Some of the messages that I sent weren‚Äôt delivered
    aa = df.iloc[:,26]
    glb.messages_not_delivered += aa.count()

    # 27 Some of the photos that I sent weren‚Äôt delivered
    ab = df.iloc[:,27]
    glb.photos_not_delivered += ab.count()

    # 28 I couldn‚Äôt play audio messages that my partners had sent to me
    ac = df.iloc[:,28]
    glb.audio_not_play += ac.count()

    # 29 The app frequently crashed or froze
    ad = df.iloc[:,29]
    glb.crashed += ad.count()

    # 30 I didn‚Äôt get notifications for new messages or chats
    ae = df.iloc[:,30]
    glb.no_notifications += ae.count()

    # 31 My messages or chats were out of order
    af = df.iloc[:,31]
    glb.orderless_messages += af.count()

    # 32 I couldn‚Äôt send audio messages to my partners
    ag = df.iloc[:,32]
    glb.audio_send_fail += ag.count()

    # 33 The video or audio call quality was really bad
    ah = df.iloc[:,33]
    glb.audio_video_quality += ah.count()

    # Other technical problem
    ai = df.iloc[:,34]
    glb.other_technical += ai.count()

    # Print total entries in dataframe
    print ("Total " + str(dataframe) + ": " + str(df["#"].count()))
```

```python
master = pd.read_excel("EN.xlsx")

languages = ["DE","ES","FR","IT","JP","KO","PT","RU","ZHS","ZHT"]

# stitch together all the files
for lang in languages:
    filename = lang + ".xlsx"
    new_file = pd.read_excel(filename)
    new_cols = {x: y for x, y in zip(new_file.columns, master.columns)}
    master = master.append(new_file.rename(columns=new_cols))
    print (master.shape)
    
# split into iOS and Android
is_android = master.iloc[:,36] == "Android"
is_ios = master.iloc[:,36] != "Android"
df_android = master[is_android]
df_ios = master[is_ios]

df_android.to_excel("df_android.xlsx", index = False)
df_ios.to_excel("df_ios.xlsx", index = False)
master.to_excel("df_concatenated.xlsx", index = False)

print ("Total Android Size:" + str(df_android.shape))
print ("Total iOS Size:" + str(df_ios.shape))
```
